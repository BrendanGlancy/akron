```python
import matplotlib.pyplot as plt
import numpy as np


citi = 11778.11 # Citi
bofa = 8185.26 # Bank of America
chase = 9104.97 # Chase
chase_cashAdvance = 827.41
consolidated = 27183.57 # consolidated
discover = 3019.61 # discover
ford_escape = 19130.94 # ford escape

# interest rate
citi_rate = 16.99
bofa_rate = 19.24
chase_rate = 22.24
chase_rate_cash = 27.99
discover_rate = 19.99
consolidated_rate = 6.75
ford_escape_rate = 2.99


# calculate the total debt
total_debt = citi + bofa + chase + chase_cashAdvance + consolidated + discover + ford_escape

# display each debt individually using matplotlib
labels = 'Citi', 'BofA', 'Chase', 'Cash Advance', 'Consolidated', 'Discover', 'Ford Escape'
sizes = [citi, bofa, chase, chase_cashAdvance, consolidated, discover, ford_escape]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'green', 'blue']
explode = (0, 0, 0, 0, 0, 0, 0)  # explode a slice if required

# Make a bar graph for the interest rates
labels2 = 'Citi', 'BofA', 'Chase', 'Advance', 'Consolidated', 'Discover', 'Car'
# Display the interest rates as a percentage
sizes2 = [citi_rate, bofa_rate, chase_rate, chase_rate_cash, consolidated_rate, discover_rate, ford_escape_rate]
colors2 = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'green', 'blue']
explode2 = (0, 0, 0, 0, 0, 0, 0)  # explode a slice if required



# Plot
plt.title('Debt Breakdown')
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()

# display the debts individually
print("Debt Breakdown")
print("---------------------------")
# Display the debt with commas
print("Citi: ${:,.2f}".format(citi))
print("Bank of America: ${:,.2f}".format(bofa))
print("Chase: ${:,.2f}".format(chase))
print("Chase Cash Advance: ${:,.2f}".format(chase_cashAdvance))
print("Consolidated: ${:,.2f}".format(consolidated))
print("Discover: ${:,.2f}".format(discover))
print("Ford Escape: ${:,.2f}".format(ford_escape))
print("---------------------------")

# Plot the interest rates on a bar graph
plt.title('Interest Rates')
plt.bar(labels2, sizes2, color=colors2)
# make the bargraph wider to fit the labels
plt.gcf().subplots_adjust(bottom=0.25)
plt.show()


# display the interest rates individually
print("Interest Rates")
print("---------------------------")
print("Citi interest rate: {}%".format(citi_rate))
print("Bank of America interest rate: {}%".format(bofa_rate))
print("Chase interest rate: {}%".format(chase_rate))
print("Chase Cash Advance interest rate: {}%".format(chase_rate_cash))
print("Consolidated interest rate: {}%".format(consolidated_rate))
print("Ford Escape interest rate: {}%".format(ford_escape_rate))
print("---------------------------\n")

print("---------------------------")
# print the total debt rounded to 2 decimal places
print("Total Debt: ${:,.2f}".format(round(total_debt, 2)))
```
